# Business Understanding

## Objective

Apply machine learning by using historical race performance matrix to predict runner performance for the NYC Marathon.

## Runner Type

Runner can be categorized by type:

amateur, moderately trained, highly trained, high-level and elite) with the common denominator that they are generally trained both in length of time and number of weekly sessions [2]



## Variables

"The most relevant variables in the marathon modality are training variables derived from the evaluation of aerobic metabolism and anthropometric parameters." [2]

Training variables: 

quantitative (mean race duration, number of training sessions per week, miles per week, km per week, training volume, miles in 8 weeks, training in 9 weeks, years of training) and qualitative (training pace, record for 1 mile, 5 miles, 10 miles, half-marathon time and having finished a marathon).

Aerobic Metabolism Assessment Variables

Vo2Max, Lactate Threshold at AT
aerobic threshold (VT1),
anaerobic threshold (VT2)

Demographic Variables
* Gender, Age, Race -> are the factors that can not be modified


### Factors influences race time

* Body Mass Composition - the leaner the faster - Ponderal index (a measure of leanness (corpulence) of a person[1] calculated as a relationship between mass and height.)[3]
* Experiences [3]
    * Especially running for the same course. As [3] indicated, the equation to predict marathon finish time, the previous race experience can be a the best prediction of a runners final time would change by 14.3 min.
* Longest training mile [3]

* Training type
    * Tempo associated with velocity for shorter distances [5]
    * typical weekly training mileage and interval training had similar associations with velocity for all race distances [5]
    * A runner who incorporates interval training should expect about a 3 % decrease in race time [5]

* Typering?
    * Any paper talk about tapering on the effect of the race performance?

## Prediction Formula
* Riegel formula [6]
    * well established, good for half marathon or shorter distance
* Slates Marathon Time Predictor [5][7]




## Problems
* Many new runners lack of proper guideline how to train for the race. 
    * " Many runners develop their training programs by imitating successful marathoners or by extrapolating from training schedules designed for shorter races. " [3]
* The training is not target a specific race, such as NYC Marathon (hilly)
* Most of the study on performance are focus on elite runners but not few study focus on recreational runner. [5]
* Recreational runner lack of special tools for effective access their race performances. [5]


## Benefit of the research

* Help new runner to determing training pace
* Help coach to modify training pace depending on the phase of the training
