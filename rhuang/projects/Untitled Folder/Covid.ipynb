```python
from covid import Covid
covid = Covid(source="john_hopkins")
covid.get_data()
```

```python
covid.get_status_by_country_name("US")
```

```python
covid.list_countries()
```

```python
import pandas as pd
df = pd.DataFrame(covid.get_data())
display(df)
```

```python
sorted = df.sort_values(by=['confirmed'], ascending=False)
sorted
```

```python
excluded = sorted[~sorted.country.isin(['Europe', 'South America', 'Asia', 'World', 'North America', 'Africa'])]
print(excluded['recovered'])
```

```python
top10 = excluded.head(10)
print(top10)
```

```python
x = top10.country
y1 = top10.confirmed
y2 = top10.active
y3 = top10.deaths
y4 = top10.recovered

y3
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

plt.figure(figsize=(5,5))
plt.plot(x, y1)
plt.xticks(rotation=90)
plt.show()
```

```python
plt.bar(x, y1)
plt.xticks(rotation=90)
plt.show()
```

```python
plt.plot(x, y1, label="Confirmed")
plt.plot(x, y2, label="Active")
plt.plot(x, y3, label="Deaths")
plt.plot(x, y4, label="Recovered")
plt.legend(loc='upper right')
plt.xticks(rotation=90)
plt.show()
```

```python
labels = ['Confirmed', 'Active', 'Deaths', 'Recovered']
plt.plot(x, y1, x, y2, x, y3, x, y4)
plt.legend(labels, loc='upper right')
plt.xticks(rotation=90)
plt.grid()
plt.show()
```

```python
df2 = pd.DataFrame([y1, y2, y3, y4])
display(x)
df2.plot.bar();
plt.legend(x, loc='upper center')
plt.xticks(rotation=90)
plt.grid()
plt.show()
```

```python
df2.plot.bar(stacked=True);
plt.legend(x, loc='upper center')
plt.xticks(rotation=90)
plt.grid()
plt.show()
```

```python
df2.plot.barh();
plt.legend(x, loc='upper center')
plt.xticks(rotation=90)
plt.grid()
plt.show()
```

```python
df2.plot.barh(stacked=True);
plt.legend(x, loc='upper center')
plt.xticks(rotation=90)
plt.grid()
plt.show()
```
